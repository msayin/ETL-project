```python
import pandas as pd
import datetime as dt
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, extract
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# Store filepath in a variable
file_team_stats = "Output/SoccerTeam_Stats.csv"
file_team_city="Output/club_city.csv"
file_city_GDP="Output/GDP by city.csv"
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
team_stats_df = pd.read_csv(file_team_stats, encoding="ISO-8859-1")
team_city_df = pd.read_csv(file_team_city, encoding="ISO-8859-1")
city_GDP_df = pd.read_csv(file_city_GDP, encoding="ISO-8859-1")
```

```python
# Preview each dataframe
team_stats_df.head()
```

```python
team_city_df.head()
```

```python
city_GDP_df.head()
```

```python
# Join team_city df with city_GDP df

Team_city_gdp_df = pd.merge(team_city_df, city_GDP_df, left_on='Location',right_on='Metropolitan areas', how="inner")
```

```python
# There are two cities named London (METRO_ID UK001 and CAN10), so we need to remove the wrong one based on METRO_ID
Team_city_gdp_df = Team_city_gdp_df[Team_city_gdp_df.METRO_ID != 'CAN10']
Team_city_gdp_df.head()
```

```python
# Join team_stats df with Team_city_gdp_df

Final_df = pd.merge(team_stats_df, Team_city_gdp_df, left_on='team_name',right_on='Club', how="inner")
```

```python
# check data types
Final_df.dtypes
```

```python
# set 2 decimals for averages
decimals = pd.Series([2, 2], index=['goal_scored_avg','goal_conceded_avg'])
Final_df=Final_df.round(decimals)
# convert total goals to integer
Final_df['goal_scored_total'] = Final_df['goal_scored_total'].apply(np.int64)
Final_df['goal_conceded_total'] = Final_df['goal_scored_total'].apply(np.int64)
Final_df.head()
```

```python
# Clean up the final columns

Final_df = Final_df[[
    'team_name',
    'country_name',
    'Location',
    'Total GDP(millions)',
    'Percapita GDP',
    'match_count',
    'goal_scored_total',
    'goal_conceded_total',
    'goal_scored_avg',
    'goal_conceded_avg',
    'dribbling',
    'dribbling_class',
    'passing',
    'passing_class',
    'speed',
    'speed_class',
    'assiting',
    'assisting_class',
    'crossing',
    'crossing_class',
    'shooting',
    'shooting_class',
    'defence',
    'defence_class']].copy()
```
